```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import file from the OS Directory

filepath = "/content/drive/MyDrive/GitHub Projects/Netflix Data Analysis/netflix_titles.csv"
```

```python
df = pd.read_csv(filepath)
df
```

```python
df.info()
```

```python
# Removing Unnecessary Columns

df.drop(['show_id', 'description'], axis = 1, inplace = True)
df
```

```python
# Checking for Dupicated Values

df.duplicated().sum()
```

```python
# Checking the null values

df.isnull().sum()
```

```python
# As there are too many null values in Director columns, We cannot remove the values.
# We can replace some values appropriately


df['director'].fillna('Unknown Director', inplace = True)
df['cast'].fillna('Unknown Cast', inplace = True)
df['country'].fillna('Unknown Country', inplace = True)
```

```python
# Checking null values again

df.isnull().sum()
```

```python
# Dropping null values

df.dropna(axis = 0, inplace = True)

# Checking null values again

df.isnull().sum()
```

```python
# Checking Shape of Dataframe

df.shape
```

```python
df.info()
```

```python
df.describe()
```

# Exploratory Data Analysis

Which type of show is mostly streaming on Netflix

```python
show_type = df['type'].value_counts()
print(show_type)
```

```python
plt.pie(show_type,labels=show_type.index,autopct='%1.1f%%')
cen_cir=plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(cen_cir)
plt.axis('equal')
plt.title('Distribution of Show Types')
plt.show()
```

Name the director who directs most of the show

```python
df['director'].value_counts().head()
```

```python
# Filtering out 'Unknown director'

director_name = df[df['director'] != 'Unknown Director']['director'].value_counts().head()
director_name
```

```python
plt.figure(figsize = (8,5))
sns.barplot(x = director_name, y = director_name.index,hue = director_name ,palette = 'Paired')
plt.xlabel("Number of Shows")
plt.ylabel("Director")
plt.title("Director with Most Shows")
plt.show()
```

Which type of shows are mostly released on Netflix based on rating

* TV-MA - Mature content
* TV-14 - Content for ages 14 and up
* TV-PG - Content for very young audiences including children for ages 2-6years
* R - Restricted for ages under 17
* PG-13 - Parental guidance for children under 13

```python
show_by_rating = df['rating'].value_counts().head()
show_by_rating
```

```python
plt.figure(figsize = (8,6))
plt.pie(show_by_rating,labels=show_by_rating.index,autopct='%1.1f%%')
plt.title("Most shows by Ratings")
plt.show()
```

Which country releases most shows on Netflix

```python
top_country = df['country'].value_counts().head()
top_country
```

```python
country = df[df['country'] != 'Unknown Country']['country'].value_counts().head()
country
```

```python
plt.figure(figsize = (8,5))
plt.pie(country,labels=country.index,autopct='%1.1f%%')
plt.title('Most shows released by Country')
plt.show()
```

Number of shows that are released in last 20 years on Netflix

```python

total_shows = df['release_year'].value_counts().sum()
print("Total shows released in last 20 years:",total_shows)
print("")
print("Year Wise releases over last 20 years: ")
shows = df['release_year'].value_counts().head(20)
print(shows)
```

```python
plt.figure(figsize = (15,6))
sns.barplot(x = shows.index, y = shows, hue = shows, palette = "Paired")
plt.title("Shows released in Last 20 years")
plt.xlabel("Years")
plt.ylabel("Number of shows Released")
plt.show()
```

Name top 5 actors

```python
actor = df['cast'].value_counts().head(10)
actor
```

```python
actor = df[df['cast'] != 'Unknown Cast']['cast'].value_counts().head(10)
actor
```

```python
plt.figure(figsize = (15,6))
sns.barplot(x = actor, y = actor.index,hue = actor.index,legend = False, palette = "bright")
plt.title("Top Actors")
plt.xlabel("Number of Shows")
plt.ylabel("Top Actors")
plt.show()
```

Name 5 latest released and oldest movies

```python
# 5 Latest released movies

latest_movie = df[df['type']=='Movie'].sort_values('release_year',ascending=False)
latest_movie[['title','release_year']].head()
```

```python
# 5 Oldest released movies\

oldest_movie = df[df['type']=='Movie'].sort_values('release_year',ascending=True)
oldest_movie[['title','release_year']].head()
```

Top categories of shows on Netflix

```python
category = df['listed_in'].value_counts().head(10)
category
```

```python
plt.figure(figsize = (8,6))
sns.barplot(x = category, y = category.index,hue = category.index, legend = False, palette = "Paired")
plt.xlabel('Categories')
plt.ylabel('Number of shows')
plt.show()
```

Most shows are added: on which date on Netflix

```python
date = df['date_added'].value_counts().head(20)
date
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x = date, y = date.index, hue = date.index, legend = False, palette = 'Paired')
plt.xlabel('Netflix added Date')
plt.ylabel('Number of shows')
plt.show()
```
